#### 1.	Check for the outliers in the “AMOUNT_IN_USD” column and replace the outliers with appropriate values, discussed in the sessions.

```python
# Specifying Q1 and Q3 quartiles and calculation lower and upper bounds and also median

Q1 = df["AMOUNT_IN_USD"].quantile(0.25)
Q3 = df["AMOUNT_IN_USD"].quantile(0.75)
lower_bound = Q1 - 1.5 * (Q3 - Q1)
upper_bound = Q3 + 1.5 * (Q3 - Q1)
median = df["AMOUNT_IN_USD"].median()

# Storing the outliers by using IQR method
outliers = df[(df["AMOUNT_IN_USD"] < lower_bound) | (df["AMOUNT_IN_USD"] > upper_bound)]

# Replacing outliers with median
df.loc[df["AMOUNT_IN_USD"] < lower_bound, "AMOUNT_IN_USD"] = median
df.loc[df["AMOUNT_IN_USD"] > upper_bound, "AMOUNT_IN_USD"] = median

df.head(3)
print(outliers["AMOUNT_IN_USD"])
```

#### Introduce a particular module for datetime manipulation. 

```python
# Filling zeroes in the values where it is of less than 6 digits
df['ORDER_CREATION_TIME'] = df['ORDER_CREATION_TIME'].astype('str').apply(lambda x: x.zfill(6))

# Converting to datetime format
creationtime = pd.to_datetime(df['ORDER_CREATION_TIME'], format='%H%M%S')

# Retrieving the hours, minutes, seconds, say, months and year of respective datetime columns and then adding those seperately to the dataset
df['ORDER_CREATION_HOUR'] = creationtime.dt.hour
df['ORDER_CREATION_MINUTE'] = creationtime.dt.minute
df['ORDER_CREATION_SECOND'] = creationtime.dt.second
df['ORDER_CREATION_DAY'] = df['ORDER_CREATION_DATE'].dt.day
df['ORDER_CREATION_MONTH'] = df['ORDER_CREATION_DATE'].dt.month
df['ORDER_CREATION_YEAR'] = df['ORDER_CREATION_DATE'].dt.year
df['REQUESTED_DELIVERY_DAY'] = df['REQUESTED_DELIVERY_DATE'].dt.day
df['REQUESTED_DELIVERY_MONTH'] = df['REQUESTED_DELIVERY_DATE'].dt.month
df['REQUESTED_DELIVERY_YEAR'] = df['REQUESTED_DELIVERY_DATE'].dt.year

df.head(3)
```

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt
```

```python
df_modified = df.copy()
lags = 5
ffday = 1
datetime_cols = 'ORDER_CREATION_DATE'
customer_id_col = 'CUSTOMER_NUMBER'
net_amount_col = 'ORDER_AMOUNT'

# Applying the difference_in_days function to the date-time columns
df_modified = difference_in_days(df_modified, lags, ffday, customer_id_col, datetime_cols, net_amount_col)

df_modified
```

# MILESTONE 4 - ML Models and Evaluations

```python
# Imports

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV, train_test_split
```

#### 1.	Modify the dataset to pass into any type of machine learning models.


```python
# Specifying the independent variables in X
X = df_modified.drop(['AMOUNT_IN_USD','ORDER_CREATION_DATE','ORDER_AMOUNT', 'REQUESTED_DELIVERY_DATE'], axis=1).astype('float')

# Specifying the dependent variables in Y
Y = df_modified.AMOUNT_IN_USD

# X.fillna(df.mean(),inplace=True)

# Splitting the variables into training and testing set where 25% of entire data is used for testing
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.25, random_state=40)
```

#### 3.	Perform Regression model evaluations like MSE, RMSE, R-Square etc.

```python
# For Linear Regression

Ypred_linear = linear_regression.predict(Xtest)

# Calculating MSE, RMSE and R-Square for testing data
mse_linear = mean_squared_error(Ytest, Ypred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(Ytest, Ypred_linear)

print("Linear Regression:")
print("MSE:", mse_linear)
print("RMSE:", rmse_linear)
print("R-Square:", r2_linear)
```

```python
# For Random Forest Regression

Ypred_rf = rf_model.predict(Xtest)

# Calculating MSE, RMSE and R-Square for testing data
mse_rf = mean_squared_error(Ytest, Ypred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(Ytest, Ypred_rf)

print("Random Forest Regression:")
print("MSE:", mse_rf)
print("RMSE:", rmse_rf)
print("R-Square:", r2_rf)
```

```python
# For AdaBoost

Ypred_ada = adaboost_model.predict(Xtest)

# Calculating MSE, RMSE and R-Square for testing data
mse_ada = mean_squared_error(Ytest, Ypred_ada)
rmse_ada = np.sqrt(mse_ada)
r2_ada = r2_score(Ytest, Ypred_ada)

print("AdaBoost:")
print("MSE:", mse_ada)
print("RMSE:", rmse_ada)
print("R-Square:", r2_ada)
```

#### 4.	Compare the accuracies of all the models 

```python
# Creatin an evaluation metrics dictionary with values from each model 
evaluation_metrics = {
    'Linear Regression': [mse_linear, rmse_linear, r2_linear],
    'Decision Tree': [mse_tree, rmse_tree, r2_tree],
    'Random Forest': [mse_rf, rmse_rf, r2_rf],
    'AdaBoost': [mse_ada, rmse_ada, r2_ada],
    'XGBoost': [mse_xgb, rmse_xgb, r2_xgb],
    # 'Support Vector Machine': [mse_svm, rmse_svm, r2_svm]
}

# Printing the evaluation metrics dictionary
print("Evaluation Metrics:")
print("{:<20} {:<15} {:<15} {:<15}".format('Model', 'MSE', 'RMSE', 'R-Square'))
for model, metrics in evaluation_metrics.items():
    mse, rmse, r2 = metrics
    print("{:<20} {:<15.2f} {:<15.2f} {:<15.11f}".format(model, mse, rmse, r2))
```

#### 5.	Select the best possible model

```python
best_model = max(evaluation_metrics, key=lambda x: evaluation_metrics[x][2])
best_model_r2 = evaluation_metrics[best_model][2]

print("The highest R-square value is {} of the {} Regression Model".format(best_model_r2, best_model))
```

#### 6.	Perform Hyperparameter tuning, select best hyperparameters by using appropriate algorithms

```python
# Defining the hyperparameters to tune

param_grid = {
    'learning_rate': [0.1, 0.01, 0.001],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 150, 200],
    'subsample': [0.8, 0.9, 1.0],
}

# Creating the XGBoost model
xgboost_model = xgb.XGBRegressor()

# Performing grid search with cross-validation
grid_search = GridSearchCV(estimator=xgboost_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(Xtrain, Ytrain)

# Storing the best hyperparameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

#### 7.	Come up with the best possible model accuracy.

```python
# Fitting values into the best model
best_model.fit(X, Y)

# Making Predictions
Ypred_best = best_model.predict(Xtest)

# Calculating and displaying various evaluation metrics
mse_best = mean_squared_error(Ytest, Ypred_best)
rmse_best = np.sqrt(mse_best)
r2_best = r2_score(Ytest, Ypred_best)

print("Best Model:")
print("MSE:", mse_best)
print("RMSE:", rmse_best)
print("R-Squared:", r2_best)
```
